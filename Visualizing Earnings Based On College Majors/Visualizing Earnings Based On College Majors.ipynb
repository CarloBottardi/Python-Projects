## Further Visualization

I'll make now use of other visualization tools.

```python
categories = recent_grads["Major_category"].unique()
print(len(categories))
```

```python
sum_male_category={}
sum_female_category={}
for c in categories:
    sum_male = recent_grads[recent_grads["Major_category"]==c]["Men"].sum()
    sum_male_category[c]=sum_male
    sum_female = recent_grads[recent_grads["Major_category"]==c]["Women"].sum()
    sum_female_category[c]=sum_female
clustered_data_male=pd.Series(sum_male_category,name="Men")
clustered_data_female=pd.Series(sum_female_category,name="Women")
```

```python
Sum_male_female=pd.concat([clustered_data_male, clustered_data_female], axis=1)
print(Sum_male_female)
```

## Box and Whiskers Plots

Useful to better visualize the values distribution of a column.
